```python
import numpy as np
import pandas as pd
from statsmodels.tsa.stattools import adfuller

# Step 1: Generate Synthetic Time Series Data
# -------------------------------------------
# Create a monthly time series of 120 months (10 years)
np.random.seed(42)  # Reproducible random results

# Create time components
date_range = pd.date_range(start='2010-01-01', periods=120, freq='M')

# Create components of time series
trend = np.linspace(50, 150, 120)  # Linearly increasing trend
seasonality = 10 * np.sin(np.linspace(0, 24, 120))  # Seasonal pattern
noise = np.random.normal(0, 5, 120)  # Random noise

# Combine to form the time series
data = trend + seasonality + noise
time_series = pd.Series(data, index=date_range)

def adfuller_test(series):
    """
    Performs the Augmented Dickey-Fuller test for stationarity.

    Args:
        series (pd.Series): The time series data to test.
    """
    print("Performing Augmented Dickey-Fuller test...")
    result = adfuller(series)
    print('ADF Statistic: %f' % result[0])
    print('p-value: %f' % result[1])
    print('Critical Values:')
    for key, value in result[4].items():
        print('\t%s: %.3f' % (key, value))

    if result[1] <= 0.05:
        print("Conclusion: The time series is likely stationary.")
    else:
        print("Conclusion: The time series is likely non-stationary.")

# Perform ADF test on the generated time series
adfuller_test(time_series)
```

## Difference the series (if needed)

### Subtask:
If the ADF test indicates non-stationarity, apply differencing to the time series to make it stationary.


**Reasoning**:
Based on the previous ADF test result (p-value > 0.05), the time series is non-stationary. I need to apply first-order differencing, drop NaN values, store the differenced series, and then perform the ADF test again to confirm stationarity.



```python
# Step 3: Difference the Series if Not Stationary
# ---------------------------------------------
# Apply first-order differencing
time_series_diff = time_series.diff().dropna()

# Perform ADF test again on the differenced series
print("\nPerforming Augmented Dickey-Fuller test on differenced series...")
adfuller_test(time_series_diff)
```

```python
# Step 4: Plot ACF and PACF to Identify p and q
# ---------------------------------------------

# Generate and display ACF plot
plot_acf(time_series_diff, lags=40)
plt.title('Autocorrelation Function (ACF)')
plt.xlabel('Lags')
plt.ylabel('Autocorrelation')
plt.show()

# Generate and display PACF plot
plot_pacf(time_series_diff, lags=40)
plt.title('Partial Autocorrelation Function (PACF)')
plt.xlabel('Lags')
plt.ylabel('Partial Autocorrelation')
plt.show()
```

## Fit arima model

### Subtask:
Based on the ACF and PACF plots, fit an ARIMA model to the (potentially differenced) time series data.

Instantiate and fit an ARIMA model to the time series data using the identified order from the ACF and PACF plots and the differencing order.



```python
# Step 5: Fit ARIMA Model
# -----------------------
# Based on the ACF and PACF plots of the differenced series:
# - ACF shows a significant spike at lag 1 and then drops off, suggesting a potential MA(1) component (q=1).
# - PACF shows a significant spike at lag 1 and then drops off, suggesting a potential AR(1) component (p=1).
# - We used first-order differencing (d=1) to make the series stationary.
# Therefore, a possible ARIMA order is (1, 1, 1).

# Instantiate the ARIMA model
# Use the original time_series data and specify the order (p, d, q)
arima_model = ARIMA(time_series, order=(1, 1, 1))

# Fit the model
arima_model_fit = arima_model.fit()

# Print the model summary
print(arima_model_fit.summary())
```

```python
# Step 5: Fit ARIMA Model
# -----------------------
# Based on the ACF and PACF plots of the differenced series:
# - ACF shows a significant spike at lag 1 and then drops off, suggesting a potential MA(1) component (q=1).
# - PACF shows a significant spike at lag 1 and then drops off, suggesting a potential AR(1) component (p=1).
# - We used first-order differencing (d=1) to make the series stationary.
# Therefore, a possible ARIMA order is (1, 1, 1).

from statsmodels.tsa.arima.model import ARIMA

# Instantiate the ARIMA model
# Use the original time_series data and specify the order (p, d, q)
arima_model = ARIMA(time_series, order=(1, 1, 1))

# Fit the model
arima_model_fit = arima_model.fit()

# Print the model summary
print(arima_model_fit.summary())
```

## Forecast

### Subtask:
Use the fitted ARIMA model to forecast the values for the next 12 time periods.


**Reasoning**:
Generate the forecast for the next 12 time periods using the fitted ARIMA model and store the results.



```python
# Step 6: Forecast the Next 12 Time Periods
# ----------------------------------------
# Use the fitted model to forecast the next 12 steps
forecast_results = arima_model_fit.forecast(steps=12)

# Print the forecasted values
print("Forecasted values for the next 12 periods:")
print(forecast_results)
```

## Plot original series and forecast

### Subtask:
Visualize the original time series and the forecasted values on a single plot to evaluate the model's performance.


**Reasoning**:
Plot the original time series and the forecasted values on a single plot to visualize the model's performance, including titles, labels, and a legend.



```python
# Step 7: Plot Original Series and Forecast
# ----------------------------------------
# Visualize the original time series and the forecasted values
plt.figure(figsize=(12, 7))
plt.plot(time_series.index, time_series, label='Original Series')
plt.plot(forecast_results.index, forecast_results, label='Forecast')
plt.title('ARIMA Forecast vs. Original Series')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```
